```python
import json
from os import environ
import requests
import time

from dotenv import load_dotenv
import polars as pl
```

```python
load_dotenv()
```

```python
def query_nyt(year: int, month: int):   
    if len(str(year)) != 4:
           raise ValueError
    
    if month not in range(0, 13):
        raise ValueError
        
    url = f"https://api.nytimes.com/svc/archive/v1/{year}/{month}.json?api-key={environ['NYT_KEY']}"
    
    return requests.get(url).json()
```

```python
articles = query_nyt(2022, 1)['response']['docs']
```

```python
kws = [{'keywords': a['keywords']} for a in articles]
```

```python
pl.from_dicts(kws)
```
